**Q1. What is an LLM, and what implications does it have for economics or finance?**

LLM은 트랜스포머 아키텍처를 기반으로 방대한 데이터를 학습하여 인간의 언어를 이해하고 추론하는 딥러닝 모델로, 금융 및 경제 분야에서 비정형 데이터의 활용 방식을 근본적으로 혁신하고 있습니다. 과거 전통적인 계량 경제 모형(OLS 회귀, ARIMA 시계열 분석 등)이 주로 과거의 주가, 거래량, 금리와 같이 이미 수치화된 정형 데이터 간의 선형적 관계를 추정하는 데 국한되었던 반면, LLM은 Self-Attention 메커니즘을 통해 단어 간의 문맥과 미묘한 뉘앙스까지 파악하여 자연어를 고차원 벡터 공간에 임베딩합니다. 이러한 인지 및 추론 로직을 통해 LLM은 연준 의사록의 매파적/비둘기파적 어조 변화, 기업 실적 발표에서 경영진이 사용하는 단어의 긍정/부정 비율, 실시간 글로벌 금융 뉴스 등 방대한 텍스트 데이터 속에서 시장의 감성과 숨겨진 알파 시그널을 즉각적인 수치로 변환하고 계량화할 수 있습니다. 이는 과거 거대 자본을 가진 기관 투자자들만이 전문 인력을 동원해 수행할 수 있었던 방대한 정성적 정보의 분석 비용과 시간을 획기적으로 단축시켜 새로운 텍스트 정보가 자산 가격에 즉시 반영되도록 이끌며, 결과적으로 개인과 기관 간의 정보 비대칭성을 크게 해소하고 시장의 준강형 효율성을 제고합니다. 나아가, 이렇게 정량화된 텍스트 데이터는 자산 가격 책정 모형의 독립적인 새로운 변수로 추가되어 전통적 재무 지표가 설명하지 못하는 시장의 심리적 움직임을 포착하게 해 주므로, 자연어 처리(NLP) 기술을 알고리즘 트레이딩에 직접 융합하여 기존의 수치 기반 모델을 뛰어넘는 차세대 퀀트 전략의 고도화를 촉진하는 중대한 파급 효과를 지닙니다.

**Q2. What are vibe coding and harness engineering, and what implications do they have for economics or finance?**

Vibe Coding은 엄격한 프로그래밍 문법 대신 자연어 프롬프트를 통해 개발자의 직관적인 의도만으로 AI가 코드를 생성하게 하는 패러다임이며, Harness Engineering은 이렇게 생성된 알고리즘의 성능, 안전성, 편향성을 체계적으로 검증하고 통제하는 평가 프레임워크를 구축하는 작업입니다. 경제 및 금융 산업에서 바이브 코딩은 프로그래밍 진입 장벽을 완전히 허물어, 단순한 코딩 구현 능력보다 투자자의 경제학적 직관과 도메인 지식 자체를 시장의 핵심 경쟁 우위로 격상시키는 중대한 영향을 미칩니다. 이를 통해 금융 분석가들은 막대한 IT 리서치 비용을 절감하면서도 딥러닝 기반의 퀀트 모델이나 알고리즘 트레이딩 전략을 신속하게 프로토타이핑하여 새로운 알파 창출 아이디어의 시장 적용 속도를 극대화할 수 있습니다. 그러나 금융 시장은 미세한 알고리즘 오류나 AI의 환각(Hallucination) 현상이 대규모 연쇄 청산이나 막대한 자본 손실과 같은 시스템적 리스크로 직결되므로 하니스 엔지니어링이 반드시 동반되어야 합니다. 하니스 엔지니어링은 엄격한 스트레스 테스트와 백테스팅을 통해 모델의 과적합을 방지하고, 금융 규제 당국이 요구하는 설명 가능한 AI(Explainable AI) 기준을 충족시키는 강력한 통제 수단으로 작용합니다. 결론적으로 바이브 코딩이 금융 혁신의 비용을 낮추고 민첩성을 높이는 '가속 페달'이라면, 하니스 엔지니어링은 리스크를 관리하고 시장의 신뢰를 방어하는 안전 장치로 기능하며, 이 둘의 상호보완적 결합이야말로 차세대 금융 시스템이 극한의 변동성 속에서도 안정적으로 작동하기 위한 필수 불가결한 인프라가 됩니다.

**Q3. In what ways does wiki-LLM improve on traditional RAG systems?**

wiki-LLM은 문서를 기계적으로 분할하여 단순한 텍스트 임베딩 간의 벡터 유사도에만 의존하는 전통적인 RAG 시스템의 정보 파편화 문제를 해결함으로써 문서 탐색 및 추론의 품질을 혁신적으로 향상시킵니다. 기존의 전통적 RAG는 연준 의사록이나 방대한 산업 리포트에서 단순히 키워드와 유사한 문장 조각들을 문맥 없이 추출하여 LLM에 주입하기 때문에, 문서 전체를 아우르는 포괄적 문맥이 소실되는 치명적인 한계가 있었습니다. 반면 wiki-LLM은 위키피디아의 하이퍼링크 구조처럼 개별 개념들 간의 계층적 관계와 의미론적 연결망을 지식 그래프 형태로 사전 구축하는 방식을 취합니다. 이러한 구조적 향상은 경제 및 금융 분야에서 필수적인 다중 홉 추론(Multi-hop reasoning), 즉 '연준의 금리 인하(A)'가 '특정 원자재 밸류체인(B)'을 거쳐 최종적으로 '개별 기업의 밸류에이션(C)'에 미치는 복잡한 연쇄적 파급 효과를 맥락의 끊김 없이 유기적으로 추적하는 것을 가능하게 합니다. 결과적으로 wiki-LLM은 단순 텍스트 짜깁기로 인해 발생하는 AI의 환각 현상을 획기적으로 억제할 뿐만 아니라, 금융 규제 및 실무에서 가장 중요한 '정보 출처의 명확한 추적성'을 보장하여 보다 신뢰도 높고 입체적인 퀀트 리서치 시스템을 구축하는 핵심 기반으로 작용합니다.

## Method 1 — Standard RAG: 4가지 질의로 검색 검증

각 논문을 타깃으로 한 영어 질문 3개 + 논문 간 교차 질문 1개를 사용한다.  
영어로 질문하는 이유: 임베딩 모델(`all-MiniLM-L6-v2`)이 영어 최적화 모델이므로  
영어 질문 시 유사도 점수가 유의미하게 높아진다.

| 질문 | 대상 문서 |
|------|-----------|
| Q1: ML models for stock return prediction | Gu et al. (2020) |
| Q2: TFT architecture and variable selection | Lim et al. (2021) |
| Q3: Fed's rate decision and economic outlook (March 2026) | FOMC 의사록 |
| Q4: Cross-paper — neural networks vs traditional methods in finance | 교차 질문 |

````python
# 4가지 질의 정의
queries = [
    {
        "id": "Q1",
        "target": "Gu et al. (2020)",
        "text": "What machine learning models are used to predict stock returns, "
                "and how do they compare to OLS regression in terms of performance?"
    },
    {
        "id": "Q2",
        "target": "Lim et al. (2021) — TFT",
        "text": "What is the architecture of the Temporal Fusion Transformer "
                "and how does it perform variable selection for time series forecasting?"
    },
    {
        "id": "Q3",
        "target": "FOMC Minutes (2026-03)",
        "text": "What was the Federal Reserve's decision on interest rates and "
                "what were the main economic risks discussed in the March 2026 FOMC meeting?"
    },
    {
        "id": "Q4",
        "target": "Cross-paper",
        "text": "How do neural networks and attention-based models improve "
                "financial forecasting compared to traditional linear methods?"
    },
]

# retrieve_only 실행 (LLM 불필요)
all_results = {}
for q in queries:
    all_results[q["id"]] = my_rag.retrieve_only(q["text"], top_k=3)

# 결과 출력 (청크 텍스트 500자까지 표시)
for q in queries:
    chunks = all_results[q["id"]]
    print("=" * 80)
    print(f"[{q['id']}] Target: {q['target']}")
    print(f"Query : {q['text']}")
    print("-" * 80)
    for i, item in enumerate(chunks):
        preview = item['chunk'].replace('\n', ' ').strip()[:500]
        print(f"  Rank {i+1} | score={item['score']:.4f} | {item['source'][:55]}")
        print(f"  └ {textwrap.fill(preview, width=76, subsequent_indent='    ')}")
    print()
````

````python
# Q1~Q4 전체 Rank·Source 종합 요약표
src_short = {
    "empircial asset pricing via macine learning.pdf": "Gu et al. (2020)",
    "Temporal Fusion Transformers for Interpretable Multi-horizon Time Series Forecasting.pdf": "Lim et al. (2021)",
    "fomcminutes20260318.pdf": "FOMC Minutes",
}

print("=" * 72)
print("[종합] Q1–Q4 retrieve_only 검색 결과 요약")
print("=" * 72)

for q in queries:
    print(f"\n  [{q['id']}] {q['target']}")
    q_text = q['text']
    print(f"  Query: {q_text[:72]}..." if len(q_text) > 72 else f"  Query: {q_text}")
    print(f"  {'Rank':>6}  {'Score':>7}  출처")
    print("  " + "-" * 50)
    for i, item in enumerate(all_results[q["id"]]):
        src = src_short.get(item['source'], item['source'][:30])
        print(f"  {'Rank '+str(i+1):>6}  {item['score']:>7.4f}  {src}")

print()
print("=" * 72)
print("[패턴 분석] 질문별 검색 출처 분포")
print("=" * 72)
print(f"  {'질문':<6}  {'Top Score':>9}  검색 출처 분포")
print("  " + "-" * 55)
for q in queries:
    chunks = all_results[q["id"]]
    sources = [src_short.get(c['source'], c['source'][:20]) for c in chunks]
    unique = list(dict.fromkeys(sources))
    top_score = chunks[0]['score']
    pattern = "단일 논문" if len(unique) == 1 else f"교차 ({', '.join(unique)})"
    print(f"  {q['id']:<6}  {top_score:>9.4f}  {pattern}")
````

## Method 1 (cont.) — Standard RAG: LLM 생성 포함 전체 파이프라인

`retrieve_only`는 검색만 한다. 아래는 검색 + LLM 생성까지 포함하는 `query()`를  
4가지 질문 각각에 실행한 결과다.

````python
# Standard RAG 전체 파이프라인: Retrieve + Generate (LLM 필요)
print("[Method 1] Standard RAG — Retrieve + Generate")
print("=" * 80)

rag_answers = {}
for q in queries:
    print(f"\n[{q['id']}] {q['target']}")
    print(f"Query: {q['text']}")
    print("-" * 80)
    try:
        result = my_rag.query(q["text"], top_k=3)
        rag_answers[q["id"]] = result.answer
        print(f"Sources    : {result.sources}")
        print(f"Time       : retrieve {result.retrieval_time:.2f}s + generate {result.generation_time:.2f}s")
        print(f"Answer     :\n{result.answer}")
    except Exception as e:
        rag_answers[q["id"]] = ""
        print(f"[오류] {type(e).__name__}: {e}")
    print()
````

## Method 2 — Karpathy LLM-Wiki

### Standard RAG와의 핵심 차이

```
Standard RAG  : 질문 → 원본 청크 검색 → LLM 답변  (지식을 매번 재발견)
Karpathy Wiki : 문서 → LLM이 위키 작성 → 위키 탐색 → LLM 답변  (지식을 한 번 컴파일)
```

### Compile Phase (최초 1회)
논문 3편 → LLM이 각각 요약 `.md` 생성 → 공통 개념 페이지 생성 → `index.md` 작성

### Query Phase
`index.md` → LLM이 관련 페이지 선택 → 위키 페이지 읽기 → 종합 답변 생성

````python
# Method 2: Karpathy LLM-Wiki
from rag_lab.karpathy_wiki import KarpathyWiki
from pathlib import Path

print("[Method 2] Karpathy LLM-Wiki")
print("=" * 80)

try:
    wiki = KarpathyWiki(papers=papers_dict, wiki_dir=Path("./wiki"), force_rebuild=False)

    # 생성된 위키 페이지 목록
    pages = wiki.list_pages()
    print(f"\n위키 페이지 ({len(pages)}개): {pages}")

    # 4가지 질문으로 위키 쿼리
    wiki_answers = {}
    for q in queries:
        print(f"\n[{q['id']}] {q['target']}")
        print(f"Query: {q['text']}")
        print("-" * 80)
        result = wiki.query(q["text"])
        wiki_answers[q["id"]] = result.answer
        print(f"탐색 페이지 수 : {result.pages_navigated}")
        print(f"Time          : retrieve {result.retrieval_time:.2f}s + generate {result.generation_time:.2f}s")
        print(f"Answer        :\n{result.answer}")

    # 위키 건강 검사
    print("\n" + "=" * 80)
    lint = wiki.lint()
    print(f"[Wiki Lint] 총 페이지: {lint['total_pages']} | 고아 페이지: {lint['orphan_pages']} | 깨진 링크: {len(lint['broken_links'])}")

except Exception as e:
    wiki_answers = {}
    print(f"[오류] {type(e).__name__}: {e}")
````
